# **I. Import Libraries**

```python
from ultralytics import YOLO
from IPython.display import Image, display
import pandas as pd
import torch
```

# **II. Model Definition**

```python
# Load model YOLOv8 pretrained (misal YOLOv8n = nano, bisa diganti v8s/v8m/v8l)
model = YOLO('yolov8n.pt')
```

`YOLO()`: Fungsi untuk memuat model YOLOv8

`yolov8n.pt`: File weights model pretrained YOLOv8n (nano version)

- `YOLOv8` memiliki beberapa ukuran model:

    - n (nano): Paling kecil, cepat, tapi akurasi lebih rendah

    - s (small)

    - m (medium)

    - l (large)

    - x (extra large): Paling besar, lambat, tapi akurasi tertinggi

# **III. Model Training**

Parameter training:

- `data='data.yaml'`: File konfigurasi yang berisi:

    - Path dataset (train/val)

    - Jumlah kelas

    - Nama-nama kelas

- `epochs=50`: Jumlah iterasi training melalui seluruh dataset

- `batch=16`: Jumlah gambar yang diproses sekaligus dalam satu batch

- `imgsz=640`: Ukuran gambar input (akan diresize ke 640x640 pixel)

- `device='0'`: Menggunakan GPU dengan ID 0 (jika tersedia)

- `patience=5`: Berhenti training jika tidak ada peningkatan val loss selama 5 epoch

```python
# Train
model.train(
    data='data.yaml',
    epochs=50, 
    batch=16,
    imgsz=640, 
    device='0', 
    patience=5          
)
```

# **IV. Model Evaluation**

## **A. Metrik Evaluasi**

```python
# Baca file hasil training
metrics = pd.read_csv('runs/detect/train/results.csv')

# Ambil metrik terakhir (epoch final)
last_epoch = metrics.iloc[-1]

# Hitung F1-Score
precision = last_epoch['metrics/precision(B)']
recall = last_epoch['metrics/recall(B)']
f1_score = 2 * (precision * recall) / (precision + recall + 1e-16)

print(f"""
METRIK UTAMA:
- Precision: {precision:.3f}
- Recall:    {recall:.3f}
- F1-Score:  {f1_score:.3f}
- mAP50:     {last_epoch['metrics/mAP50(B)']:.3f}
- mAP50-95:  {last_epoch['metrics/mAP50-95(B)']:.3f}
""")
```

### **3. Performa Metrik**

**F1 - Score**

Penggunaan F1 score dalam model YOLO untuk deteksi objek Personal Protective Equipment (PPE) pada industri makanan sangat penting karena metrik ini memberikan keseimbangan antara precision (ketepatan) dan recall (kelengkapan), yang keduanya krusial dalam konteks keselamatan dan akurasi. Dalam deteksi PPE seperti apron, sarung tangan, masker, dan hairnet, precision memastikan bahwa model tidak salah mengklasifikasikan objek non-PPE sebagai PPE, yang dapat menyebabkan kepercayaan yang salah terhadap keamanan pekerja. Sementara itu, recall memastikan bahwa semua instance PPE yang ada dideteksi, sehingga tidak ada pelanggaran protokol keselamatan yang terlewat. F1 score, sebagai rata-rata harmonik dari keduanya, menjadi metrik ideal karena lingkungan makanan memerlukan akurasi tinggi untuk mencegah kontaminasi, di mana kesalahan baik berupa false positive maupun false negative dapat berdampak serius pada kesehatan konsumen. Dengan demikian, F1 score membantu mengevaluasi performa model secara holistik, memastikan model YOLO dapat diandalkan untuk aplikasi real-time di industri makanan.

```python
display(Image(filename='runs/detect/train/BoxF1_curve.png', width=900)) 
```

Kurva gabungan "all classes" dengan nilai F1 maksimum 0,87 pada confidence 0,412 menunjukkan performa keseluruhan yang baik, dengan `Apron`  dan `Glove` tampak mendominasi kontribusi positif. Namun, beberapa kategori seperti `No_Hairnet` dan `No_Mask` menunjukkan F1 yang lebih rendah, menandakan tantangan dalam mendeteksi ketidakhadiran item tertentu. 

Secara keseluruhan, model ini memiliki kinerja yang solid dengan pengaturan confidence optimal di sekitar 0,4, meskipun ada ruang untuk perbaikan pada deteksi kategori yang lebih spesifik.

# **VI. Kesimpulan**

Berdasarkan analisis keseluruhan dari confusion matriks, grafik loss train dan validasi, serta kurva F1-Confidence, model YOLO untuk deteksi Personal Protective Equipment (PPE) dalam industri makanan menunjukkan performa yang cukup baik. Confusion matriks menunjukkan prediksi yang akurat pada kategori seperti No_Glove (317) dan Apron (178), meskipun terdapat kesalahan klasifikasi pada kategori seperti Hairnet dan background. Kurva F1-Confidence mencapai nilai maksimum 0,87 pada confidence 0,412, dengan kontribusi utama dari kategori seperti apron dan glove, meskipun deteksi pada kategori "No" (seperti No_Hairnet) masih perlu ditingkatkan.

## **A. Pengembangan**

1. `Peningkatan Data Latih`: Tambahkan data latih khusus untuk kategori dengan performa rendah seperti No_Hairnet dan No_Mask untuk meningkatkan recall dan mengurangi false negatives.

2. `Penyesuaian Confidence Threshold`: Optimalkan ambang batas kepercayaan di sekitar 0,4-0,5 berdasarkan kurva F1 untuk menyeimbangkan precision dan recall pada semua kelas.

3. `Augmentasi Data`: Terapkan teknik augmentasi seperti rotasi, skala, dan perubahan pencahayaan untuk meningkatkan robustitas model terhadap variasi kondisi di lingkungan kerja.

4. `Hyperparameter Tuning`: Eksperimen dengan parameter seperti learning rate atau ukuran batch untuk memperhalus konvergensi loss, terutama pada dfl loss yang menunjukkan fluktuasi kecil.

5. `Evaluasi Tambahan`: Gunakan metrik seperti mAP (mean Average Precision) pada berbagai IoU threshold untuk mendapatkan gambaran lebih komprehensif tentang performa model di berbagai skenario.

Dengan pendekatan ini, model dapat diperbaiki untuk mencapai akurasi yang lebih tinggi dan reliabilitas yang lebih baik dalam mendeteksi PPE secara real-time di industri makanan.
